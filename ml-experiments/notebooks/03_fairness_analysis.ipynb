# ML Experiments: Fairness and Demographic Distribution Analysis

This notebook evaluates allocation fairness and representation metrics across candidates.

```python
import pandas as pd
import matplotlib.pyplot as plt

report_df = pd.read_csv('../datasets/fairness_report.csv')
print("=== Demographic Metrics Summary ===")
print(report_df)
```

## 1. Rural vs Urban Representation

```python
geo_df = report_df[report_df['Dimension'] == 'Rural/Urban']
print(geo_df)

plt.bar(geo_df['Group'], geo_df['Average_Match_Score'], color=['green', 'blue'])
plt.title('Average Score by Geography')
plt.ylabel('Score %')
plt.ylim(0, 100)
plt.show()
```

## 2. Caste Category Distribution

```python
cat_df = report_df[report_df['Dimension'] == 'Category']
print(cat_df)

plt.bar(cat_df['Group'], cat_df['Average_Match_Score'], color='orange')
plt.title('Average Score by Caste Category')
plt.ylabel('Score %')
plt.ylim(0, 100)
plt.show()
```
