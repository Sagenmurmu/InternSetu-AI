# ML Experiments: Recommendation Engine Validation and Testing

This notebook demonstrates matches retrieval, scoring breakdown, and natural-language explanations verification.

```python
import pandas as pd
import matplotlib.pyplot as plt

matches_df = pd.read_csv('../datasets/match_results_sample.csv')
candidates_df = pd.read_csv('../datasets/candidates_sample.csv')
internships_df = pd.read_csv('../datasets/internships_sample.csv')

print(f"Loaded {len(matches_df)} match evaluations.")
```

## 1. Score Distributions

```python
print(matches_df['final_score'].describe())

plt.hist(matches_df['final_score'], bins=20, edgecolor='black')
plt.title('Distribution of Match Final Scores')
plt.xlabel('Final Score %')
plt.ylabel('Frequencies')
plt.show()
```

## 2. Inspecting Top Recommendations for Sample Candidate

```python
sample_cand_id = 1
cand_matches = matches_df[matches_df['candidate_id'] == sample_cand_id].sort_values(by='final_score', ascending=False).head(5)

cand_info = candidates_df[candidates_df['candidate_id'] == sample_cand_id].iloc[0]
print(f"=== Candidate: {cand_info['name']} ===")
print(f"Skills: {cand_info['skills']}")
print(f"Sector Interest: {cand_info['sector_interest']}\n")

print("=== Top 5 Recommendations ===")
for idx, row in cand_matches.iterrows():
    job_info = internships_df[internships_df['internship_id'] == row['internship_id']].iloc[0]
    print(f"Internship #{job_info['internship_id']}: {job_info['title']} at {job_info['company_name']}")
    print(f"  - Final Score: {row['final_score']}% (Skills: {row['skill_score']}%, Qual: {row['qualification_score']}%, Location: {row['location_score']}%)")
    print(f"  - Sector Interest: {job_info['sector']}")
    print(f"  - Required Skills: {job_info['required_skills']}\n")
```
