# ML Experiments: Candidate and Internship Data Analysis

This notebook provides exploratory data analysis on the generated synthetic candidate profiles and internship postings.

```python
import pandas as pd
import json
import matplotlib.pyplot as plt

candidates_df = pd.read_csv('../datasets/candidates_sample.csv')
internships_df = pd.read_csv('../datasets/internships_sample.csv')

print(f"Loaded {len(candidates_df)} candidates and {len(internships_df)} internships.")
```

## 1. Candidate Demographic Analysis

```python
# Gender Breakdown
print("=== Gender Distribution ===")
print(candidates_df['gender'].value_counts())

# Category Breakdown
print("\n=== Category Distribution ===")
print(candidates_df['category'].value_counts())

# Rural vs Urban
print("\n=== Geography Distribution ===")
print(candidates_df['rural_or_urban'].value_counts())
```

## 2. Candidate Qualifications and Academic Levels

```python
print("=== Qualifications Distribution ===")
print(candidates_df['qualification'].value_counts())
```

## 3. Top Candidate Skills and Sector Interests

```python
skills_list = []
for sk in candidates_df['skills']:
    try:
        skills_list.extend(json.loads(sk))
    except Exception:
        pass

skills_series = pd.Series(skills_list)
print("=== Top 10 Candidate Skills ===")
print(skills_series.value_counts().head(10))

print("\n=== Sector Interests ===")
print(candidates_df['sector_interest'].value_counts().head(5))
```

## 4. Internship Demand and Capacity Breakdown

```python
print("=== Capacity Utilization statistics ===")
total_capacity = internships_df['capacity'].sum()
total_selected = internships_df['selected_count'].sum()
print(f"Total Capacity: {total_capacity}")
print(f"Total Selected: {total_selected}")
print(f"Platform Vacancy Rate: {(total_capacity - total_selected)/total_capacity * 100:.2f}%")
```
